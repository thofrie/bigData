```python
def preprocessing(df):
    df.date = pd.to_datetime(df.date)
    df.datecreated = pd.to_datetime(df.datecreated)
    df = df.drop(columns=["deliverydatepromised","deliverydatereal"])
    df = df.drop(columns=["points"])
    df = df.drop(columns=["w4"])
    df = df.drop(columns=["customernumber"])
    df = df.drop(columns=["invoicepostcode","delivpostcode"])
    return df
df = preprocessing(df)
```
